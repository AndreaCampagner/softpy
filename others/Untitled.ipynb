```python
import numpy as np
```

```python
class Comparable:
    def __init__(self, val1, val2):
        self.val1 = val1
        self.val2 = val2

    def __eq__(self, other):
        return self.val1 == other.val1 and self.val2 == other.val2

    def __lt__(self, other):
        return (self.val1 <= other.val1 and self.val2 < other.val2) or (self.val1 < other.val1 and self.val2 <= other.val2)

    def __gt__(self, other):
        return (self.val1 >= other.val1 and self.val2 > other.val2) or (self.val1 > other.val1 and self.val2 >= other.val2)

    def __ge__(self, other):
        return self.val1 >= other.val1 and self.val2 >= other.val2

    def __str__(self):
        return str(self.val1) + " " + str(self.val2)
```

```python
def max(input_list):
    output_list = []
    for i in input_list:
        check = True
        for j in input_list:
            if i < j:
                check = False
                break
        if check:
            output_list.append(i)
    return output_list
```

```python
l = [Comparable(1,2), Comparable(1,1), Comparable(2,1)]
```

```python
print(np.max(l))
```

```python
for i in max(l):
    print(i)
```
