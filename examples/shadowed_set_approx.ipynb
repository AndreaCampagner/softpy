```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append("../source/")
from fuzzy.fuzzyset import *
from fuzzy.operations import *
import fuzzy.approximations as approx
```

```python
# Example usage
elements = [1, 2, 3, 4, 5, 6, 7]
membership = [0.2, 0.3, 0.6, 0.8, 0.5, 0.1, 0.4]

# fuzzy_set = FuzzySet(elements, membership)
fuzzy_set = DiscreteFuzzySet(elements, membership)

alpha_range = np.linspace(0, 1, num=100)
```

```python
def plot_and_print_sets(fuzzy_set, shadowed_set, best_alpha):
    print(shadowed_set.memberships, "membership")
    print(best_alpha, "best_alpha")
    
    print("Original Fuzzy Set:")
    print("Elements:", fuzzy_set.items)
    print("Membership:", fuzzy_set.memberships)
    
    print("\nShadowed Set with Best Alpha ({}):".format(best_alpha))
    print("Elements:", shadowed_set.items)
    print("Membership:", shadowed_set.memberships)
    
    
    # Original Fuzzy Set
    plt.subplot(2, 1, 1)
    plt.plot(fuzzy_set.items, fuzzy_set.memberships, 'bo-', label='Original Fuzzy Set')
    plt.xlabel('Elements')
    plt.ylabel('Membership Grade')
    plt.title('Original Fuzzy Set')
    plt.legend()
    
    # Shadowed Set
    plt.subplot(2, 1, 2)
    plt.plot(fuzzy_set.items, fuzzy_set.memberships, 'bo-', label='Original Fuzzy Set')
    plt.plot(shadowed_set.items, shadowed_set.memberships, 'ro-', label='Shadowed Set')
    # plt.axvline(x=best_alpha, linestyle='--', color='gray', label='Threshold (alpha)')
    plt.axhline(y=best_alpha, linestyle='--', color='gray', label='Threshold (alpha)')
    plt.xlabel('Elements')
    plt.ylabel('Membership Grade')
    plt.title('Shadowed Set')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
# si pu scegliere tra 'gradualness_balance', 'sharpness_balance' e 'tradeoff'
shadowed_set, best_alpha = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'gradualness_balance')
plot_and_print_sets(fuzzy_set, shadowed_set, best_alpha)
```

```python
# si pu scegliere tra 'gradualness_balance', 'sharpness_balance' e 'tradeoff'
shadowed_set, best_alpha = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'sharpness_balance')
plot_and_print_sets(fuzzy_set, shadowed_set, best_alpha)
```

```python
# si pu scegliere tra 'gradualness_balance', 'sharpness_balance' e 'tradeoff'
shadowed_set, best_alpha = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'tradeoff')
plot_and_print_sets(fuzzy_set, shadowed_set, best_alpha)
```

```python
# Example usage for operations
elements = [1, 2, 3, 4, 5, 6, 7]
membership = [0.2, 0.3, 0.6, 0.8, 0.5, 0.1, 0.4]

# fuzzy_set = FuzzySet(elements, membership)
fuzzy_set = DiscreteFuzzySet(elements, membership)

alpha_range = np.linspace(0, 1, num=100)
```

```python
shadowed_set, best_alpha = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'gradualness_balance')
shadowed_set2, best_alpha2 = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'sharpness_balance')
shadowed_set3, best_alpha3 = approx.shadowed_set_induction(fuzzy_set, alpha_range, 'tradeoff')
```

```python
#applying the product operation on shadowed sets (shadowed sets 1 and 2)
result = sh_op.product(shadowed_set, shadowed_set2)
print("Product operation on shadowed sets \nShadowed Set 1:")
print("Elements:", shadowed_set.items)
print("Membership:", shadowed_set.memberships, "\n")

print("Shadowed Set 2:")
print("Elements:", shadowed_set2.items)
print("Membership:", shadowed_set2.memberships, "\n")

print("Result Set:")
print("Elements:", result.items)
print("Membership:", result.memberships, "\n")


#applying the lukasiewicz operation on shadowed sets (shadowed sets 2 and 3)
result = sh_op.lukasiewicz(shadowed_set2, shadowed_set3)
print("Lukasiewicz operation on shadowed sets \nShadowed Set 2:")
print("Elements:", shadowed_set2.items)
print("Membership:", shadowed_set2.memberships, "\n")

print("Shadowed Set 3:")
print("Elements:", shadowed_set3.items)
print("Membership:", shadowed_set3.memberships, "\n")

print("Result Set:")
print("Elements:", result.items)
print("Membership:", result.memberships, "\n")
```
